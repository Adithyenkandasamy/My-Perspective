```python
# Funtion for Chatting with the Installed Model
import requests

def call_llm(input):
  response = requests.post(
      "http://127.0.0.1:11434/api/generate",
      json={
          "model": model_name,
          "prompt": input,
          "stream": False
          }
  )
  return response
```

```python
import time

user_input = input("Enter the message: ")

start_time = time.time()

response = call_llm(user_input)

end_time = time.time()

total_time = end_time - start_time

data = response.json()

print(data)

if "response" in data:
    print("\nAI:", data["response"])
else:
    print("\nNo response received")

print("\nSeconds:", total_time)
print("Milliseconds:", total_time * 1000)
```

```python
# FastAPI Text-to-Text Model API

from fastapi import FastAPI
from pydantic import BaseModel

import requests
import uvicorn
import threading

app = FastAPI()

# Mention your model name here
model_name = "mistral"


# Input format
class ChatRequest(BaseModel):

    prompt: str


@app.post("/chat")

async def chat_api(data: ChatRequest):

    response = requests.post(
        "http://127.0.0.1:11434/api/generate",

        json={

            "model": model_name,

            "prompt": data.prompt,

            "stream": False
        }
    )

    result = response.json()

    return {
        "response": result.get("response", "No response")
    }


# Run FastAPI
def run():

    uvicorn.run(
        app,
        host="0.0.0.0",
        port=8000
    )


threading.Thread(target=run).start()
```
