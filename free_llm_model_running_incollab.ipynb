```python
# Model installing code for the Ollama Mention the Model name in the name area
import requests
import json


'''
> Knowledge:
- llama3
- mistral
- gemma2:2b

>Coding Models:
- qwen2.5-coder:7b
- deepseek-coder:6.7b
- codellama:7b

>Faster Models:
- phi3:mini
- tinyllama
- gemma2:2b
'''

# Mention the model name Here
model_name = "llama3"

response = requests.post(
    "http://127.0.0.1:11434/api/pull",
    json={
        "name": model_name # Mention the Model name
    },
    stream=True
)

for line in response.iter_lines():
    if line:
        print(json.loads(line))
```

```python
import time

user_input = input("Enter the message: ")

start_time = time.time()

response = call_llm(user_input)

end_time = time.time()

total_time = end_time - start_time

data = response.json()

print(data)

if "response" in data:
    print("\nAI:", data["response"])
else:
    print("\nNo response received")

print("\nSeconds:", total_time)
print("Milliseconds:", total_time * 1000)
```

```python
# Funtion for Chatting with the Installed Model
import requests

def call_llm(input):
  response = requests.post(
      "http://127.0.0.1:11434/api/generate",
      json={
          "model": model_name,
          "prompt": input,
          "stream": False
          }
  )
  return response
```
