```python
# FastAPI OCR Server

from fastapi import FastAPI, UploadFile, File
import requests
import base64
import uvicorn
import threading

app = FastAPI()

# Mention your OCR model here
model_name = "maternion/lightonocr-2:1b"


@app.post("/ocr")

async def ocr_api(file: UploadFile = File(...)):

    # Read image
    image_bytes = await file.read()

    # Convert to base64
    image_base64 = base64.b64encode(
        image_bytes
    ).decode("utf-8")

    # Send image to Ollama
    response = requests.post(
        "http://127.0.0.1:11434/api/generate",

        json={

            "model": model_name,

            "prompt": "Extract all text from this image and return raw text only.",

            "images": [image_base64],

            "stream": False
        }
    )

    data = response.json()

    return {
        "ocr_result": data.get("response", "No response")
    }


# Run FastAPI
def run():

    uvicorn.run(
        app,
        host="0.0.0.0",
        port=8000
    )


threading.Thread(target=run).start()
```
